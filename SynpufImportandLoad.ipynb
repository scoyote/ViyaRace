```python
def loadSynpuf(path,filename,sample,suffix):
    fn = filename+str(sample)+suffix+".zip"
    cn = filename+str(sample)+suffix+".csv"
    print(fn)
    url = path+fn
    http = urllib3.PoolManager()
    urllib3.disable_warnings()
    response = http.request('GET',url)
    with open(fn,'wb') as f:
        f.write(response.data)
    zip_ref = zipfile.ZipFile(fn, 'r')
    zip_ref.extractall("/home/viyauser/SamsViya/ViyaRace")
    zip_ref.close()
    return sess.upload(cn).casTable

def cleanUp():
    for filename in os.listdir("./"):
        if filename.endswith(".csv") or filename.endswith("zip"):
            os.remove(filename)
```
